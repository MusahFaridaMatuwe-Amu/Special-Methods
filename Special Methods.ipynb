 Python Special Methods Reference Guide

Special methods are often called **dunder methods** because they start and end with double underscores. They allow you to connect your custom classes directly to Python's built-in syntax. By defining these methods, you make your own objects behave exactly like native Python lists, numbers, or functions.#%% md


### 1. Constructor (__init__)
* **What it is:** This is the initialization method. It automatically runs the very moment you create a new object from a class.
* **Why we use it:** To set up the initial properties (attributes) of your object right at the start.


```python
class SmartDevice:
    def __init__(self, name, device_type):
        self.name = name
        self.device_type = device_type
        self.is_on = False  # Default state

# Instantiating the object triggers __init__ automatically
phone = SmartDevice("iPhone", "Smartphone")
print(f"Device: {phone.name}, Type: {phone.device_type}, Status: {phone.is_on}")
```

### 2. String Representation (__str__, __repr__)
* **What it is:** These methods control how your object looks when it is converted into text. `__str__` is for a clean, user-friendly description, while `__repr__` is a detailed, technical description meant for developers.
* **Why we use it:** To prevent Python from printing ugly, unreadable memory addresses (like `<__main__.Book object at 0x7f...>`) when you view your object.


```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f"'{self.title}' by {self.author}"

    def __repr__(self):
        return f"Book(title='{self.title}', author='{self.author}')"

my_book = Book("The Hobbit", "J.R.R. Tolkien")

print(str(my_book))   # Triggers __str__
print(repr(my_book))  # Triggers __repr__
```

### 3. Length (__len__)
* **What it is:** This method links your custom object directly to Python's built-in `len()` function.
* **Why we use it:** To allow anyone using your class to easily count how many items, elements, or data entries are stored inside your object.


```python
class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song_title):
        self.songs.append(song_title)

    def __len__(self):
        return len(self.songs)

my_playlist = Playlist("Chill Vibes")
my_playlist.add_song("Song A")
my_playlist.add_song("Song B")

# Using the built-in len() function on our custom object
print(f"Playlist length: {len(my_playlist)}")
```

### 4. Equality (__eq__)
* **What it is:** This method defines how the double equals operator (`==`) compares two objects from your class.
* **Why we use it:** By default, Python only checks if two objects occupy the exact same spot in computer memory. We use this method to tell Python to look inside and compare the actual data values instead.


```python
class Product:
    def __init__(self, item_id, price):
        self.item_id = item_id
        self.price = price

    def __eq__(self, other):
        if isinstance(other, Product):
            return self.item_id == other.item_id
        return False

prod1 = Product("P100", 25.0)
prod2 = Product("P100", 25.0)

print(prod1 == prod2)  # True, because their item_ids match
```

### 5. Operator Overloading (__add__, __sub__)
* **What it is:** These methods allow your custom objects to interact naturally with basic math symbols like addition (`+`) and subtraction (`-`).
* **Why we use it:** To let you mathematically combine or subtract objects intuitively without having to write clunky, custom method names like `.add_together()`.


```python
class Wallet:
    def __init__(self, cash):
        self.cash = cash

    def __add__(self, other):
        if isinstance(other, Wallet):
            return Wallet(self.cash + other.cash)
        return NotImplemented

    def __sub__(self, other):
        if isinstance(other, Wallet):
            return Wallet(self.cash - other.cash)
        return NotImplemented

    def __repr__(self):
        return f"Wallet(${self.cash})"

wallet1 = Wallet(50)
wallet2 = Wallet(30)

print("Combined:", wallet1 + wallet2)
print("Remaining:", wallet1 - wallet2)
```

### 6. Item Access (__getitem__)
* **What it is:** This method lets you look up and read specific pieces of data from your object using standard square bracket indexing notation (like `object[index]`).
* **Why we use it:** To make your custom object behave exactly like a native Python list or dictionary for quick data retrieval.


```python
class CustomTeam:
    def __init__(self):
        self.players = ["Alice", "Bob", "Charlie"]

    def __getitem__(self, index):
        return self.players[index]

team = CustomTeam()

# Accessing elements using square brackets []
print(team[0])
print(team[2])
```

### 7. Item Assignment (__setitem__)
* **What it is:** This method lets you change, update, or assign a value inside your object using square brackets and an equals sign (like `object[index] = value`).
* **Why we use it:** To give users a clean, familiar way to modify items inside your custom collections directly, without needing a separate update method.


```python
class Inventory:
    def __init__(self):
        self.slots = ["Empty", "Empty", "Empty"]

    def __setitem__(self, index, value):
        self.slots[index] = value

    def __repr__(self):
        return f"Inventory: {self.slots}"

bag = Inventory()
bag[0] = "Shield"  # Modifies index 0 directly
bag[1] = "Potion"  # Modifies index 1 directly

print(bag)
```

### 8. Membership Testing (__contains__)
* **What it is:** This method handles the background logic when you use the `in` keyword to check if an item exists inside a collection.
* **Why we use it:** It allows you to write clean, conversational conditional expressions like `if item in object:` instead of manually writing loop searches.


```python
class Warehouse:
    def __init__(self):
        self.stock = ["Laptop", "Monitor", "Keyboard"]

    def __contains__(self, item):
        return item in self.stock

hub = Warehouse()

print("Laptop" in hub)       # Returns True
print("Smartphone" in hub)   # Returns False
```

### 9. Callable Objects (__call__)
* **What it is:** This method turns an instance of your class into something that can be executed directly using parentheses `()`, exactly like a standard function.
* **Why we use it:** To create objects that behave like functions but are still able to store and remember their own internal data across multiple uses.


```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, number):
        return number * self.factor

# Create a triple multiplier instance
triple = Multiplier(3)

# Call the object directly like a function
result = triple(10)
print(f"Result: {result}")
```

### 10. Iteration Support (__iter__, __next__)
* **What it is:** These twin methods turn your object into an iterable sequence. `__iter__` starts the loop process, and `__next__` hands over the data pieces one by one until it runs out.
* **Why we use it:** To safely plug your custom objects directly into standard Python `for` loops so they can be stepped through sequentially.


```python
class Countdown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration  # Tells the loop to stop cleanly
        self.current -= 1
        return self.current + 1

# Testing iteration in a standard loop
counter = Countdown(3)
for num in counter:
    print(num)
```

### 11. Boolean Evaluation (__bool__)
* **What it is:** This method dictates whether an instance of your class is interpreted as `True` (truthy) or `False` (falsy) when it is evaluated in a logical test.
* **Why we use it:** To control the behavior of your objects inside `if` statements, allowing you to define what makes an object "empty" or "active".


```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __bool__(self):
        # Account is True if it has money, False if empty or overdrawn
        return self.balance > 0

acc = BankAccount(0)

if not acc:
    print("Action required: This account is empty!")
```

### 12. Destructor (__del__)
* **What it is:** This is the cleanup method. It runs automatically right before Python completely removes an object from system memory.
* **Why we use it:** To release external resources that the object was holding onto—such as closing database sessions, network connections, or file handles.


```python
class FileSimulator:
    def __init__(self, filename):
        self.filename = filename
        print(f"Opening connection to {self.filename}")

    def __del__(self):
        print(f"Closing connection and freeing memory for {self.filename}")

# Create the object
session = FileSimulator("data.csv")

# Manually delete the object to trigger the destructor immediately
del session
```
